```python
most_versatile = { 'orifice_size': 75,
 'aspect_ratio': 1,
 'expansion_ratio': 4.5,
 'normalized_orifice_length': 3,
 'normalized_water_inlet': 2,
 'normalized_oil_inlet': 2
 }
least_versatile = { 'orifice_size': 150,
 'aspect_ratio': 3,
 'expansion_ratio': 2,
 'normalized_orifice_length': 1,
 'normalized_water_inlet': 4,
 'normalized_oil_inlet': 4
 }
```

```python
for i, key in enumerate(least_versatile.keys()):
    if i == 0:
        least_results = all_results.loc[all_results.loc[:, key] == least_versatile[key],:]
    else:
        least_results = least_results.loc[least_results.loc[:, key] == least_versatile[key],:]
        
```

```python
MOST_CHIPNUM = 629
LEAST_CHIPNUM = 21495
most_results = most_results.loc[most_results.regime == 1, :]
least_results = least_results.loc[least_results.regime == 1, :]
most_least_chips = chips.iloc[[MOST_CHIPNUM, LEAST_CHIPNUM], :]
```

```python
most_least_chips = most_least_chips.drop("Unnamed: 0 ", axis=1)
```

```python
highest_results = most_results
lowest_results = least_results
print("LOWEST %f" % most_least_chips.tail(1)["score"])
print("HIGHEST %f" % most_least_chips.head(1)["score"])


for x in [lowest_results, highest_results]:
    sizes = x["droplet_size"].tolist()
    rates = x["generation_rate"].tolist()
    points = np.array([[sizes[i], rates[i]] for i in range(len(sizes))])
    hull = ConvexHull(points)
    plt.scatter(points[:,0], points[:,1], s=10)
    for simplex in hull.simplices:
        plt.plot(points[simplex, 0], points[simplex, 1], 'k-')
        
plt.title("Design Space Coverage of Most and Least Versatile Chips")
plt.xlabel("Droplet Size (um)")
plt.ylabel("Generation Rate (Hz)")
#plt.savefig("VersatilityExtremes_Compare.png")
```

```python
all_filt = all_results.loc[all_results["diff"] < 50, :]
drip_filt = all_filt.loc[all_filt["regime"] == 1, :]
jet_filt = all_filt.loc[all_filt["regime"] == 2, :]
```

```python
score = []
size_score = []
rate_score = []
hulls = []
for i, chip in tqdm(enumerate(chips.iterrows()), total=chips.shape[0],position=0, leave=True):
    chip_data = all_filt.loc[all_filt.loc[:, "chip_num"] == i, :]
    sizes = np.array(chip_data.loc[:, "droplet_size"])
    rates = np.array(chip_data.loc[:, "generation_rate"])
    points = np.array([[sizes[i], rates[i]] for i in range(len(sizes))])
    try:
        hull = ConvexHull(points)
        hulls.append(hull)
        size_score.append(np.max(sizes) - np.min(sizes))
        rate_score.append(np.max(rates) - np.min(rates))
        score.append(hull.volume)
    except:
        hulls.append(-1)
        score.append(-1)
        size_score.append(-1)
        rate_score.append(-1)
chips.score = score
chips.size_score = size_score
chips.rate_score = rate_score
```
